```python
import requests
from datetime import datetime

# https://www.football-data.org/client/home
# https://footballtickets-by-gakuseimiler.com/entry/football-data-api

# APIキー
api_key = ""
headers = {"X-Auth-Token": api_key}

# APIからデータ取得
url = "https://api.football-data.org/v2/competitions/PL/matches"
response = requests.get(url, headers=headers)
data_CL = response.json()

# JSON取得後の処理
game_list = data_CL["matches"]
games_num = len(game_list)
youbi = ["日", "月", "火", "水", "木", "金", "土"]

for i in range(games_num):
    # 日時を日本時間に変換
    utc_date = datetime.strptime(game_list[i]["utcDate"], "%Y-%m-%dT%H:%M:%SZ")
    jdate = utc_date.strftime("%Y/%m/%d(%a)")
    jtime = utc_date.strftime("%H:%M")

    # 日程表にデータを挿入
    row = (
        f'<tr align="center">'
        f'<td><img src="https://crests.football-data.org/{game_list[i]["homeTeam"]["id"]}.svg" height="30">'
        f'<br />{game_list[i]["homeTeam"]["name"]}</td>'
        f'<td class="{game_list[i].get("td_class", "")}">{jdate}<br />{jtime}</td>'
        f'<td><img src="https://crests.football-data.org/{game_list[i]["awayTeam"]["id"]}.svg" height="30">'
        f'<br />{game_list[i]["awayTeam"]["name"]}</td>'
        f'</tr>'
    )
    # 表示または保存などの処理を行う場合に適切なコードを追加

    print(row)
```
